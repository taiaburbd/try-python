```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#image show using color matplotlib
img=mpimg.imread("images/monkey.jpg")

plt.imshow(img)
plt.colorbar()
```

```python
#image using cv2

color_imgcv2=cv2.imread("images/monkey.jpg",1)
grey_imgcv2=cv2.imread("images/monkey.jpg",0)

cv2.imshow("RGB Image",color_imgcv2)
cv2.imshow("Grey Image",grey_imgcv2)

cv2.waitKey(1000)
cv2.destroyAllWindows()
```

```python
#image display by bgr to rgb
plt.imshow(color_imgcv2)
```

```python
#BGR to RGB manual convert
#cv2 image format BGR and matplotlib format RGB

bgr2rgb=color_imgcv2[...,::-1]
plt.imshow(bgr2rgb)
```

```python
#BGR to RGB cv2 convert
plt.imshow(cv2.cvtColor(color_imgcv2,cv2.COLOR_BGR2RGB))
```
