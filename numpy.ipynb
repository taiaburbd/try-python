```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,num=100)
y=np.cos(x)

plt.plot(x,y)
plt.show()
```

f : [−3, 3] → R defined by f(x) = e−x2

```python
def f(x):
    return np.exp(-x**2)

a,b=-3,3
x=np.linspace(a,b,num=100)
y=f(x)
plt.plot(x,y)
plt.savefig('exp.png')
plt.show()
```

Plotting functions point by point

```python
x=np.array([0,1,2,3,4,5])
y=np.array([0,1,4,1,3,2])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linewidth=3,color='red')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

Multiple routes
f k(x) = 1/(1 + e−kx)

```python
def f(k,x):
    return 1/(1+np.exp(-k*x))

a,b=-5,5
x=np.linspace(a,b,num=100)

for k in range(1,5):
    y=f(x,k)
    plt.plot(x,y,label="k={}".format(k))

plt.legend()
plt.show()
```
