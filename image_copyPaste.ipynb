```python
from PIL import Image
import matplotlib.pyplot as plt

img_rgb=Image.open('images/rgb.jpeg')
img_monkey= Image.open('images/monkey.jpg')
img_monkey.thumbnail((110,200))
# rgb image copy
img_new =img_rgb.copy()
#axial plane mid of img_rgb
axial_plane=int((img_rgb.size[1]-img_monkey.size[1])/2)
max_value=img_rgb.size[0]
# how many monkey can seat
step=int(img_rgb.size[0]/int(img_rgb.size[0]/(img_monkey.size[0])))
start=step-img_monkey.size[0]

# single line monkey sitting
for i in range(start,max_value,step):
    img_new.paste(img_monkey,(i,axial_plane))
#     print(i,step,axial_plane)
# print(img_rgb.size)
# print(img_monkey.size)

plt.imshow(img_new)
```
